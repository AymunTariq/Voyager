```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import pandas as pd
```

```python
data=pd.read_csv('trial_data.csv')
df_senate= data.copy()
```

```python
df_senate.shape
```

```python
df_senate.set_index('S_no',inplace=True)
```

```python
df_senate.head()
```

```python
X = df_senate
```

```python
X.head()
```

```python
ward = AgglomerativeClustering(n_clusters=3)
ward_pred = ward.fit_predict(X)
```

```python
# Hierarchical clustering using complete linkage
# Write the code to create an instance of AgglomerativeClustering with the appropriate parameters
complete = AgglomerativeClustering(n_clusters=3,linkage='complete')

# Fit & predict
# Write a code to make AgglomerativeClustering fit the dataset and predict the cluster labels
complete_pred = complete.fit_predict(X)

# Hierarchical clustering using average linkage
# Write the code to create an instance of AgglomerativeClustering with the appropriate parameters
avg = AgglomerativeClustering(n_clusters=3,linkage='average')

# Fit & predict
# Write a code to make AgglomerativeClustering fit the dataset and predict the cluster labels
avg_pred = avg.fit_predict(X)
```

```python
# Import scipy's linkage function to conduct the clustering
from scipy.cluster.hierarchy import linkage

#Write code here
linkage_type = 'complete' 

linkage_matrix = linkage(X,linkage_type)
```

```python
# write the code to plot using 'dendrogram()'
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
abc = plt.figure(figsize=(15,18))
dendrogram(linkage_matrix,leaf_rotation=0,leaf_font_size=10,labels=X.index,orientation='right')
plt.show()
```
